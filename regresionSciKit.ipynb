# Regresión lineal con scikit-learn

```
from sklearn.feature_selection import RFE
from sklearn.svm import SVR
import pandas as pd
import numpy as np
```

```
data = pd.read_csv("https://raw.githubusercontent.com/joanby/python-ml-course/master/datasets/ads/Advertising.csv")
```

```
feature_cols = ["TV", "Radio", "Newspaper"]
```

```
X = data[feature_cols]
Y = data["Sales"]
#Y es la variable a predecir.
```

```
estimator = SVR(kernel="linear")
selector = RFE(estimator, 2, step=1) #Recursive feature elimination.
#Me quiero quedar con dos variables y que lo haga en un paso.
selector = selector.fit(X,Y)
```

Información y parámetros del SVR https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVR.html

```
selector.support_
#Descarta "Newspaper".
```

```
selector.ranking_
#A menor número, más representativa/significativa.
```

```
#Hacer un modelo lineal con scikit
from sklearn.linear_model import LinearRegression
```

```
X_pred = X[["TV", "Radio"]]
```

```
lm = LinearRegression()
lm.fit(X_pred, Y)
```

```
lm.intercept_
```

```
lm.coef_
```

```
lm.score(X_pred,Y)
```
