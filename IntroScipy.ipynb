```
import numpy as np
import pandas as pd

import scipy.stats as stats
```

```
pesos = [100, 150, 150, 200, 250, 300, 325, 400, 415, 500, 600, 1000]
```

```
#Media
pesos_media = np.mean(pesos)
pesos_media
```

```
#Mediana
np.median(pesos)
```

```
#Moda
stats.mode(pesos)
```

```
#Rango
np.max(pesos) - np.min(pesos)
#Diferencia entre el mayor y el menor
```

```
#Cuartiles e IQR
stats.mstats.mquantiles(pesos)
```

```
#La función pandas.qcut convierte una lista de N cuantiles. Si le pasamos 4, tenemos los cuartiles. 
#qcut devuelve una serie categórica nueva con los cuantiles como categorías.
pesos_cuartiles = pd.qcut(pesos, 4)
pesos_cuartiles.categories
```

Aquí se ve que los intervalos que separan los pesos en 4 bloques son:<br>
Categories(4, interval[float64]): [(99.999, 187.5] < (187.5, 312.5] < (312.5, 436.25] < (436.25, 1000]]
<br>
Es importante notar que hay diversas formas de obtener los cuartiles, de aquí la diferencia con el método calculado a mano.

```
#Desviación standard
pesos_std = np.std(pesos)
pesos_std
```

Coeficiente de simetría <br>
El coeficiente de simetría (skewness) se define como el tercer momento dividido por la desviación standard al cubo. 

```
pow3 = lambda x: x*x*x
```

```
tercer_momento = lambda x: pow3(x - pesos_media)
```

```
simetria_pesos = sum(map(tercer_momento, pesos)) / (12*pow3(pesos_std))
simetria_pesos
#Map funciona como el apply de R para evitar hacer un for
```

```
#Lo mismo, pero precargado
stats.skew(pesos)
```

```
pow4 = lambda x: x*x*x*x
cuarto_momento = lambda x: pow4(x - pesos_media)
curtosis_pesos = sum(map(cuarto_momento, pesos)) / (12*pow4(pesos_std)) - 3
curtosis_pesos
```

```
stats.kurtosis(pesos)
```
