```
def inversador(frase):
  frase = frase.split(" ")
  fraseB = frase[::-1]
  fraseB = " ".join(fraseB)
  return fraseB
  
print(inversador("Esta es una frase"))
```

```
def popularizador(frase):
  b = [letra for letra in frase if letra not in " ,.\n"]
  popularLetra = Counter(b).most_common(2)
  return popularLetra
frase = "El veloz murcielago hindu comia algo"
a = popularizador(frase)
print(a)
```

```
def Jaccard(grupo1, grupo2):
  union = len(set(grupo1).union(set(grupo2)))
  interseccion = len(set(grupo1).intersection(grupo2))
  return interseccion / union
grupo_amigos1 = ["Manfredo", "Godofredo", "Lázaro", "Alfredo"]
grupo_amigos2 = ["Lázaro", "Tiburcio", "Egberto", "Enrico"]

jaccard_amigos = Jaccard(grupo_amigos1, grupo_amigos2)
print(jaccard_amigos)
```

Hacer una función que dado el nombre de un archivo, lo lea y devuelva la línea con la mayor longitud.

```
import os
def linea_mas_larga(nombre):
  with open(nombre) as fname:
    lineas = [linea.strip("\n") for linea in fname.readlines()]
  linea_mas_larga = lineas[0]
  for linea in lineas:
    if len(linea) > len(linea_mas_larga):
      linea_mas_larga = linea
  return linea_mas_larga
linea_mas_larga("C:/DataScience08012019/textos.txt")
```

Hacer una función que tenga como argumento el nombre de un archivo y un número n y devuelva las n últimas líneas.

```
def leer_n_ultimas(nombre, n):
  with open(nombre, "r") as fname:
    lineas = [linea.strip("\n") for linea in fname.readlines()]
  return lineas[-n:]
```
